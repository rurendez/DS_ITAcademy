### Level 2 - Exercise 4
* Show different examples of arrays, on the rule of broadcasting tells "the arrays can be broadcasted if their dimensions match or if one it the arrays has a shape of 1 

Broadcasting describes how Numpy treats arrays with different shapes during arithmetic operations. The idea is to broadcast the smaller array in order to have compatible shapes.

When the 2 arrays have the same shape the operation can be done without any problem

```python
import numpy as np

# Define 2 arrays with the same shape
a = np.array([[5,7,9],[5,7,8]])
b = np.array([[3,4,5],[7,8,9]])

# Arithmetical operation
c = a*b 

# All are ofthe same shape
print("Shape array a: ", a.shape)
print("Shape array b: ", b.shape)
print("Shape array c: ", c.shape)
```

In case the arrays do not have the same shape, it will be added one dimension of 1 to the beginning of the array with less dimensions. 

```python
import numpy as np

# Define a 3 dimension array, shape (1,1,5)
a = np.arange(5).reshape(1,1,5)
print("Shape array a: ", a.shape)
print("Array a: ",a)

# Define a 1 dimension array, shape (5,) 
b= np.arange(5)
print("Shape array b: ", b.shape)
print("Array b: ", b)

# Sum both arrays
c=a+b
print("Array c: ", c)

#  It has been  possible to sum both arrays due that numpy has applied the 
# broadcast rule, it has changed the shape of b array form (5,) to (1,1,5)
newb = b.reshape(1,1,5)
print("Array newb: ", newb)

# As we can see with the newb array the result of the operation is the same as in c array
newc = a + newb
print("Array newc: ", newc)
```

This also happen when we add an scalar to an array, The scalar has 0 dimension, son Numpy has to apply the Broadcast rule and add dimensions to the scalar. 

```python
import numpy as np

# Define a 3 dimension array, shape (1,1,5)
a = np.arange(5).reshape(1,1,5)
print("Shape array a: ", a.shape)
print("Array a: ",a)

# Define a scalar, 0 dimension 
b= 10
print("Scalar b: ",b)

# Sum both arrays
c=a+b
print("Array c: ", c)
#  It has been  possible to sum both arrays due that numpy has applied the 
# broadcast rule, it has changed the shape of the scalar form 0-D to (1,1,5),
# where the value 10 has been streched
```

If the array has a value of 1 on any of its dimensions, it will behave as it has the greatest length in that dimension.

```python
import numpy as np

# Define a 2 dimension array, shape (2,3)
a = np.arange(6).reshape(2,3)
print("Shape array a: ", a.shape)
print("Array a: ",a)

# Define a 2 dimension array, shape (2,1)
b = np.array([[50],[60]])
print("Shape array b: ", b.shape)
print("Array b: ",b)

# Sum both arrays
c=a+b
print("Shape array c: ", c.shape)
print("Array c: ",c)

#  In this case Numpy has broadcasted array b, it order to have 
# [[50,50,50],[60,60,60]]
```

Once applied the rules of Broadcasting, if the dimension of the arrays does not match, the arithmetical operation between arrays can not be done.

```python
import numpy as np

# Define a 2 dimension array, shape (2,3)
a = np.arange(6).reshape(2,3)
print("Shape array a: ", a.shape)
print("Array a: ",a)

# Define a 1 dimension array, shape (2,)
b = np.array([5,6])
print("Shape array b: ", b.shape)
print("Array b: ",b)

#  In this case Numpy can not apply neither of the rules, so the sum operation
# can not be completed

# Let's check if there is any error
try:
    c=a+b
    print("The sum is ok!")
except:
    print("The sum can not be done")
```

### Level 2 - Exercise 5
* Using array indexing, extract the values of one row, one column, and sum the values

```python
import numpy as np

# Array definition
arr = np.array([[20, 22, 43], [45, 35, 40]])
num_rows, num_cols = arr.shape

myPrint ("Total rows: ", num_rows)
myPrint ("Total columns: ", num_cols)

# Extract values of row 1 -> [0] and sum them
tot = 0
val=[]
for v in range (num_cols):
    val.append(arr[0][v])
    tot += arr[0][v]
myPrint("\nValues 1st Row: ", val)
myPrint("Sum Values 1st Row: ", tot)

# Extract values of Column 3 -> [2] and sum them
tot = 0
val=[]
for v in range (num_rows):
    val.append(arr[v][2])
    tot += arr[v][2]
myPrint("\nValues 3rd Column: ", val)
myPrint("Sum values of 3rd Column: ", tot)

# Sum of the elemnts of the array using indexing
tot = 0
for r in range (num_rows):
    for c in range (num_cols):
        tot += arr[r][c]
myPrint("\nSum of all elemnts of the array: ", tot)
```
